```python
# Check for null values
print("Null Values:")
print(df.isnull().sum())
```

```python
# Summary statistics of the data
df.describe().transpose()
```

```python
# Fill missing values with a specific value
df['Sub-Category'].fillna('phones', inplace=True)
```

```python
# Check for null values again
print("Null Values:")
print(df.isnull().sum())
```

```python
# Check for duplicate rows based on all columns
duplicate_rows = df[df.duplicated()]
duplicate_rows
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
df
```

```python
#Save the Cleaned Data
df.to_csv('cleaned_data.csv', index=False)
```

```python
# Time Analysis
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Shipping Time'] = (df['Ship Date'] - df['Order Date']).dt.days
print("Average Shipping Time:", df['Shipping Time'].mean())
```

```python
# Segment Analysis
segment_sales = df.groupby('Segment')['Price Per Each'].sum()
print("Segment-wise Sales:\n", segment_sales)
```

```python
# Product Analysis
category_sales = df.groupby('Category')['Price Per Each'].sum()
print("Category-wise Sales:\n", category_sales)
```

```python

# Discount Analysis
discount_sales = df.groupby('Discount')['Price Per Each'].sum()
print("Discount-wise Sales:\n", discount_sales)
```

```python
# Geographic Analysis
region_sales = df.groupby('Region')['Price Per Each'].sum()
print("Region-wise Sales:\n", region_sales)


```

```python
# Shipping Mode Analysis
ship_mode_sales = df.groupby('Ship Mode')['Price Per Each'].sum()
print("Ship Mode-wise Sales:\n", ship_mode_sales)
```

```python

# Customer Analysis
top_customers = df.groupby('Customer ID')['Price Per Each'].sum().nlargest(5)
print("Top 5 Customers:\n", top_customers)
```

```python

# Correlation Analysis
correlation_matrix = df.corr()
print("Correlation Matrix:\n", correlation_matrix)
```

```python
# Visualization
sns.barplot(x='Segment', y='Price Per Each', data=df)
plt.title('Segment-wise Sales')
plt.show()
```

```python

sns.barplot(x='Category', y='Price Per Each', data=df)
plt.title('Category-wise Sales')
plt.show()
```

```python

sns.scatterplot(x='Discount', y='Price Per Each', data=df)
plt.title('Discount vs Sales')
plt.show()
```

```python
# Bar Plot for Region-wise Sales
plt.subplot(2, 2, 1)
sns.barplot(x='Region', y='Price Per Each', data=df, palette="viridis")
plt.title('Region-wise Sales')
```

```python
# Box Plot for Shipping Time
plt.subplot(2, 2, 2)
sns.boxplot(x='Segment', y='Shipping Time', data=df, palette="mako")
plt.title('Shipping Time by Segment')
```

```python
# Pair Plot for Numerical Variables
plt.subplot(2, 2, 3)
sns.pairplot(df[['Quantity', 'Discount', 'Buy Price', 'Price Per Each']], palette="plasma")
plt.suptitle('Pair Plot for Numerical Variables', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Heatmap for Correlation Matrix
plt.subplot(2, 2, 4)
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix Heatmap')
plt.tight_layout()
plt.show()
```
